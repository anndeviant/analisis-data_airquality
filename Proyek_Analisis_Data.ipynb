*Selanjutnya ialah mengatasi outlier dengan winsorization menggunakan percentile (membagi data ke 100 bagian)*

```python
def apply_winsorizing(df, lower_percentile=0.05, upper_percentile=0.95):
    df_winsorized = df.copy()
    # Memilih kolom numerik kecuali kolom yang ingin dikecualikan
    numeric_columns = df.select_dtypes(include='number').columns
    # Menghapus kolom yang tidak perlu di-winsorized ('No', 'RAIN', dan 'year' dikecualikan)
    numeric_columns = numeric_columns.difference(['No', 'RAIN', 'year'])
    for column in numeric_columns:
        lower_limit = df[column].quantile(lower_percentile)  # Menggunakan percentiles
        upper_limit = df[column].quantile(upper_percentile)  # Menggunakan percentiles
        df_winsorized[column] = df[column].clip(lower=lower_limit, upper=upper_limit)
    return df_winsorized

# Menerapkan Winsorization pada setiap DataFrame
df_Aotizhongxin_winsorized = apply_winsorizing(df_Aotizhongxin_merged)
df_Changping_winsorized = apply_winsorizing(df_Changping_merged)
df_Dingling_winsorized = apply_winsorizing(df_Dingling_merged)
df_Dongsi_winsorized = apply_winsorizing(df_Dongsi_merged)
df_Guanyuan_winsorized = apply_winsorizing(df_Guanyuan_merged)
df_Gucheng_winsorized = apply_winsorizing(df_Gucheng_merged)
df_Huairou_winsorized = apply_winsorizing(df_Huairou_merged)
df_Nongzhanguan_winsorized = apply_winsorizing(df_Nongzhanguan_merged)
df_Shunyi_winsorized = apply_winsorizing(df_Shunyi_merged)
df_Tiantan_winsorized = apply_winsorizing(df_Tiantan_merged)
df_Wanliu_winsorized = apply_winsorizing(df_Wanliu_merged)
df_Wanshouxigong_winsorized = apply_winsorizing(df_Wanshouxigong_merged)

def apply_rain_outlier_removal(df):
    df_rain_cleaned = df.copy()
    rain_threshold = 1.0  # Nilai threshold untuk menghilangkan outlier
    df_rain_cleaned['RAIN'] = df_rain_cleaned['RAIN'].apply(lambda x: x if x <= rain_threshold else 0)
    # Tambahkan kondisi untuk menghilangkan nilai outlier yang lebih besar dari 72.5
    df_rain_cleaned['RAIN'] = df_rain_cleaned['RAIN'].apply(lambda x: x if x <= 72.5 else 0)
    return df_rain_cleaned

# Menerapkan fungsi outlier removal pada setiap DataFrame
df_Aotizhongxin_winsorized = apply_rain_outlier_removal(df_Aotizhongxin_winsorized)
df_Changping_winsorized = apply_rain_outlier_removal(df_Changping_winsorized)
df_Dingling_winsorized = apply_rain_outlier_removal(df_Dingling_winsorized)
df_Dongsi_winsorized = apply_rain_outlier_removal(df_Dongsi_winsorized)
df_Guanyuan_winsorized = apply_rain_outlier_removal(df_Guanyuan_winsorized)
df_Gucheng_winsorized = apply_rain_outlier_removal(df_Gucheng_winsorized)
df_Huairou_winsorized = apply_rain_outlier_removal(df_Huairou_winsorized)
df_Nongzhanguan_winsorized = apply_rain_outlier_removal(df_Nongzhanguan_winsorized)
df_Shunyi_winsorized = apply_rain_outlier_removal(df_Shunyi_winsorized)
df_Tiantan_winsorized = apply_rain_outlier_removal(df_Tiantan_winsorized)
df_Wanliu_winsorized = apply_rain_outlier_removal(df_Wanliu_winsorized)
df_Wanshouxigong_winsorized = apply_rain_outlier_removal(df_Wanshouxigong_winsorized)

# Menampilkan hasil
print(df_Aotizhongxin_winsorized.head())
print(df_Aotizhongxin_winsorized.isna().sum().sum())
max_rain_aotizhongxin = df_Aotizhongxin_winsorized['RAIN'].max()
print(f'Maksimum curah hujan (RAIN) di Aotizhongxin: {max_rain_aotizhongxin}')
maxyear = df_Aotizhongxin_winsorized['year'].max()
print(f'Maksimum year di Aotizhongxin: {maxyear}')
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana perbandingan tingkat PM2.5 dan PM10 rata-rata antar kota dan kota mana yang memiliki tingkat polusi PM2.5 dan PM10 tertinggi dan terendah?

```python
df_clean['date_time'] = pd.to_datetime(df_clean['date_time'])

rata_rata_pm25 = df_clean.groupby('station')['PM2.5'].mean()
rata_rata_pm10 = df_clean.groupby('station')['PM10'].mean()

rata_rata = pd.DataFrame({'PM2.5': rata_rata_pm25, 'PM10': rata_rata_pm10})

rata_rata = rata_rata.sort_values(by='PM2.5', ascending=False)

plt.figure(figsize=(15, 10))
rata_rata.plot(kind='bar', alpha=0.7)
plt.title('Rata-rata Tingkat PM2.5 dan PM10 di Setiap Kota')
plt.xlabel('Kota')
plt.ylabel('Rata-rata Tingkat (µg/m³)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.legend(title='Jenis Partikel')
plt.show()

kota_pm25_tertinggi = rata_rata_pm25.idxmax()
kota_pm25_terendah = rata_rata_pm25.idxmin()

kota_pm10_tertinggi = rata_rata_pm10.idxmax()
kota_pm10_terendah = rata_rata_pm10.idxmin()

# Mencetak hasil
print(f"Kota dengan rata-rata tingkat PM2.5 tertinggi adalah {kota_pm25_tertinggi}")
print(f"Kota dengan rata-rata tingkat PM2.5 terendah adalah {kota_pm25_terendah}")
print(f"Kota dengan rata-rata tingkat PM10 tertinggi adalah {kota_pm10_tertinggi}")
print(f"Kota dengan rata-rata tingkat PM10 terendah adalah {kota_pm10_terendah}")
```

### Pertanyaan 2: Bagaimana tren tahunan tingkat rata-rata SO2 di berbagai kota dari 2013 hingga 2017, dan apakah ada pola umum yang terlihat?

```python
df_clean['date_time'] = pd.to_datetime(df_clean['date_time'])

yearly_so2 = df_clean.groupby([df_clean['date_time'].dt.year, 'station'])['SO2'].mean().unstack()

# Supaya tahun tetap integer
yearly_so2 = yearly_so2[yearly_so2.index.astype(int).isin(yearly_so2.index)]

# Membuat plot rata-rata tingkat SO2
plt.figure(figsize=(15, 10))
yearly_so2.plot(marker='o')
plt.title('Rata-rata Tingkat SO2 per Tahun untuk Setiap Kota')
plt.xlabel('Tahun')
plt.ylabel('Rata-rata Tingkat SO2')
plt.xticks(yearly_so2.index.astype(int), rotation=45)  # Memastikan sumbu x menunjukkan tahun integer
plt.legend(title='Kota', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Menghitung dan mencetak perubahan keseluruhan tingkat SO2 untuk setiap kota
for city in yearly_so2.columns:
    total_change = yearly_so2[city].iloc[-1] - yearly_so2[city].iloc[0]
    print(f"Perubahan keseluruhan tingkat SO2 untuk {city} dari tahun {yearly_so2.index[0]} ke {yearly_so2.index[-1]}: {total_change:.2f}")
```

Grafik menunjukkan tren penurunan yang signifikan pada tingkat SO2 di semua kota yang diamati selama periode 2013 hingga 2016. Hal ini mengindikasikan keberhasilan upaya dalam mengurangi emisi sulfur dioksida. Penurunan ini kemungkinan besar disebabkan oleh penerapan regulasi yang lebih ketat terkait emisi industri dan kendaraan bermotor, peralihan ke sumber energi yang lebih bersih, serta peningkatan kesadaran masyarakat akan pentingnya kualitas udara. Meski demikian, terdapat perbedaan tingkat penurunan di setiap kota, yang menunjukkan bahwa efektivitas upaya pengendalian polusi udara bervariasi antar wilayah dan dipengaruhi oleh faktor-faktor seperti struktur industri, kepadatan penduduk, dan kondisi geografis.

### Pertanyaan 3: Bagaimana pola musiman tingkat NO2 di berbagai kota dan kota mana yang mengalami fluktuasi NO2 paling signifikan antar musim di Negara Beijing, China?

```python
df_clean['date_time'] = pd.to_datetime(df_clean['date_time'])

# Adding season column based on Chinese seasons
def get_season(month):
    if month in [3, 4, 5]:
        return 'Musim Semi'
    elif month in [6, 7, 8]:
        return 'Musim Panas'
    elif month in [9, 10, 11]:
        return 'Musim Gugur'
    else:
        return 'Musim Dingin'

df_clean['musim'] = df_clean['date_time'].dt.month.map(get_season)

# Calculating average NO2 levels for each season and city
rata_rata_musim_no2 = df_clean.groupby(['station', 'musim'])['NO2'].mean().unstack()

plt.figure(figsize=(14, 8))
sns.heatmap(rata_rata_musim_no2, annot=True, fmt='.1f', cmap='YlOrRd')
plt.title('Rata-rata Tingkat NO2 Berdasarkan Musim dan Kota di Beijing, Tiongkok')
plt.ylabel('Kota')
plt.tight_layout()
plt.show()

# Calculating seasonal variation for each city
variasi_musim = rata_rata_musim_no2.max(axis=1) - rata_rata_musim_no2.min(axis=1)
kota_variasi_terbesar = variasi_musim.idxmax()
print(f"Kota dengan variasi musiman NO2 terbesar: {kota_variasi_terbesar}")
```
